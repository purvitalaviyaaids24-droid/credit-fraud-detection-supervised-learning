# Credit Card Fraud Detection Summary Report

## Business Problem

Credit card fraud causes significant financial losses. Since fraud transactions are very rare compared to genuine transactions, accuracy alone is not a reliable evaluation metric. A model predicting every transaction as legitimate can achieve high accuracy while completely failing to detect fraud.

## Imbalance Handling

To address class imbalance, three approaches were tested: the original dataset, SMOTE oversampling, and random undersampling. SMOTE produced the best balance between fraud detection and false positives, improving recall without losing too much precision.

## Best Model

Several machine learning models were evaluated, including Logistic Regression, Random Forest, and XGBoost. After hyperparameter tuning, XGBoost achieved the highest PR-AUC and F1-score. Threshold optimization further improved fraud detection by selecting the threshold that maximized the F1-score instead of using the default threshold of 0.5.

## Business Impact

A business cost-benefit analysis was performed by assuming an average fraud amount of ₹4,500 and an investigation cost of ₹150 per flagged transaction. The optimized threshold produced the highest net benefit by increasing fraud detection while controlling investigation costs. This demonstrates that business metrics should be considered alongside machine learning metrics.

## Future Improvements

In a production environment, the model should support real-time transaction scoring, automatic retraining using new fraud data, continuous model drift monitoring, and analyst feedback to improve future predictions. These improvements would help maintain high fraud detection performance as fraud patterns evolve.
